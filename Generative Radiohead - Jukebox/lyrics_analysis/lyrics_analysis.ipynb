```python
import re
import os
from collections import Counter
import random
```

```python
directory_path = './lyrics'
all_lyrics = []
count = 0

for dirpath, dirnames, filenames in os.walk(directory_path):
    for file_name in filenames:

        file_full_path = os.path.join(dirpath, file_name)
        with open(file_full_path, 'r') as file:            
            lyrics = file.read()
            lyrics = lyrics.lower()
            lyrics = re.sub(r'\W+', ' ', lyrics)
            words = lyrics.split()
            all_lyrics.extend(words)
  
word_counts = Counter(all_lyrics)
print(word_counts)        
```

```python
top_num = 200
wordPair = word_counts.most_common(top_num)
gen_lyrics = ""
count = 0
line_count = 0

for _ in range(top_num):
    
    line_len = random.randint(4, 4)
    
    word = random.choice(wordPair)
    
    if count % line_len == 0 and count > 0:
        gen_lyrics += word[0] + " \n"
        
        if line_count % 24 == 0 and count > 0:
            gen_lyrics += "\n"
        
        count = 0
    else:
        gen_lyrics += word[0] + " "
        
   
        
    count +=1
    line_count +=1
    
print(gen_lyrics)
```
